unique va in1d, setdiff1d

```python
sonlar=np.array([12,23,34,3,1,2,1,22,5.6,6,6,34])
np.unique(sonlar)
```

```python
raqamlar=np.array([12,23,6])
np.in1d(sonlar,raqamlar)
```
